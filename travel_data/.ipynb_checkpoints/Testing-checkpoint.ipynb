```python
travel_details['Accommodation cost'] = travel_details['Accommodation cost'].replace(['500 USD'],500)
```

```python
def replace_price(price):
    price = str(price)
    price = price.replace('USD',"")
    price = price.replace('$',"")
    price = price.replace(',',"")
    return price
```

```python
travel_details["Accommodation cost"] = travel_details["Accommodation cost"].apply(lambda x : replace_price(x))
```

```python
top_5_cities = (travel_details['Destination'].value_counts()).to_frame().reset_index().head(5)
type(top_5_cities)
```

```python
top_5_cities
```

```python
dictionary_cities = top_5_cities.to_dict('records')
dictionary_cities
```

```python
sns.barplot(data=top_5_cities, x=top_5_cities['Destination'], y=top_5_cities['count'])
plt.xticks(rotation=45)
plt.xlabel('CITY')
plt.ylabel('SUM')
plt.title('TOP 5 CITIES')
```
